```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import tensorflow as tf
from scipy.stats import norm
from sklearn.mixture import GaussianMixture as GMM
from scipy import stats
import json
```

```python
window = 20

def get_dict_from_file(file_name):
    """
    Reads in dictionary of H values.
    """
    with open('h_dict_'+'w_{}_'.format(window)+file_name, 'r') as file:
        h_dict = file.read()
    
    return json.loads(h_dict)
```

```python
filename_control = 'Control_frame001-200'
h_dict_control = get_dict_from_file(filename_control)
keys_list = list(h_dict_control)
plt.figure()
plt.hist(h_dict_control[keys_list[0]], 20, density=True, histtype='step', label='step size 1')
#plt.hist(h_dict_control[keys_list[3]], 25, density=True, histtype='step', label='step size 4')
plt.hist(h_dict_control[keys_list[4]], 20, density=True, histtype='step', label='step size 5')
plt.xlabel('H')
plt.ylabel('p(H)')
plt.title(filename_control)
plt.legend()

filename_mutant = 'LanB1_frame001-200'
h_dict_mutant = get_dict_from_file(filename_mutant)
keys_list = list(h_dict_mutant)
plt.figure()
plt.hist(h_dict_mutant[keys_list[0]], 20, density=True, histtype='step', label='step size 1')
#plt.hist(h_dict_mutant[keys_list[3]], 25, density=True, histtype='step', label='step size 4')
plt.hist(h_dict_mutant[keys_list[4]], 20, density=True, histtype='step', label='step size 5')
plt.xlabel('H')
plt.ylabel('p(H)')
plt.title(filename_mutant)
plt.legend()
```
