### Code to calculate mean squared displacement and persistence 

```python
import numpy as np
import pandas as pd
import json
import os
import matplotlib.pyplot as plt
```

```python
def get_disp_dict_from_file(filename, window, step_size, max_step_size, restriction):

    with open('disp_dict_data/'+filename+'_s{}/'.format(max_step_size)+'disp_dict_'+filename
              +'_w{}_step{}_r{}'.format(window, step_size, restriction), 'r') as f:
        disp_dict = f.read()
    
    return json.loads(disp_dict)

def get_persistence_track(displacement_arr):
    
    inx = np.array([(displacement_arr[1:]-displacement_arr[0:-1])])
    
    path_length = sum(abs(inx))
    
    return displacement_arr[-1]/path_length
    
def get_persistence_from_dict(disp_dict):
    # for one step size

    persistence_arr = np.array([])
    
    keys_list = list(disp_dict)  # keys are 'TrackID[downsampling start index]'
    length = len(keys_list)
    for i, ds_tid in enumerate(keys_list):
        print('Track: {}   ({}/{})'.format(ds_tid, i+1, length))
        disps = np.array(disp_dict[ds_tid])
        
        p = get_persistence_track(disps)
        persistence_arr = np.append(persistence_arr,p)

    return persistence_arr
"""
def save_p_data_single_step(p_data_single_step, filename, window, step_size, max_step_size, restriction):
    # save h data for one step size (and all tracks) to csv file
    
    path = 'p_data/'+filename+'_s{}/'.format(max_step_size)
    try:
        np.savetxt(path+'p_data_'+filename+'_w{}_step{}_r{}'.format(window, step_size, restriction), 
                   p_data_single_step, delimiter=',')
    except FileNotFoundError:
        os.makedirs(path)
        np.savetxt(path+'p_data_'+filename+'_w{}_step{}_r{}'.format(window, step_size, restriction), 
                   p_data_single_step, delimiter=',')
"""       
def get_p_dict(disp_dict, step_sizes, window, restriction):
    """
    """
    p_dict = {}
    
    for i, s in enumerate(step_sizes):
        p_arr = get_persistence_from_dict(disp_dict)
        p_dict["{}".format(s)] = p_arr.tolist()
        
    return p_dict

def save_p_dict(p_dict, file_name, step_sizes, window, restriction):
    """
    """
    with open('p_dict_'+'w{}_'.format(window)+'s{}_'.format(max(step_sizes))+'r{}_'.format(restriction)+file_name, 'w') as file:
        file.write(json.dumps(p_dict))

def main(filenames, window, step_sizes, restriction):

    for f in filenames:
        print(f)
        for s in step_sizes:
            print('Step size: {}'.format(s))
            disp_dict = get_disp_dict_from_file(f, window, s, max(step_sizes), restriction)
            persistence_dict = get_p_dict(disp_dict,step_sizes, window, restriction)
            save_p_dict(persistence_dict,f,step_sizes,window,restriction)
            #save_p_data_single_step(persistence_data, f, window, s, max(step_sizes), restriction)
            del persistence_dict
            #gc.collect()
```
